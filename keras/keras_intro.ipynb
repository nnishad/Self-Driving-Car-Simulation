```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
n_pts=500
np.random.seed(0)

Xa=np.array([np.random.normal(10,2,n_pts),np.random.normal(12,2,n_pts)]).T
Xb=np.array([np.random.normal(5,2,n_pts),np.random.normal(6,2,n_pts)]).T
X=np.vstack((Xa,Xb))
Y=np.matrix(np.append(np.zeros(n_pts),np.zeros(n_pts))).T

plt.scatter(X[:n_pts,0],X[:n_pts,1])
plt.scatter(X[n_pts:,0],X[n_pts:,1])
```

```python
model=Sequential()
model.add(Dense(units=1,input_shape=(2,),activation='sigmoid'))
adam=Adam(lr=0.1)
model.compile(adam,loss='binary_crossentropy',metrics=['accuracy'])
h=model.fit(x=X,y=Y,verbose=1,batch_size=50,epochs=500,shuffle='true')
```

```python
plt.plot(h.history['acc'])
plt.title('accuracy')
plt.xlabel('epoch')
plt.legend(['accuracy'])
```

```python
plt.plot(h.history['loss'])
plt.title('loss')
plt.xlabel('epoch')
plt.legend(['accuracy'])
```

```python
def boundary(X,Y,model):
    x_span=np.linspace(min(X[:,0])-1,max(X[:,0])+1,50)
    y_span=np.linspace(max(Y[:,0])-1,max(X[:,0])+1,50)
    xx,yy=np.meshgrid(x_span,y_span)
    xx_,yy_=xx.ravel(),yy.ravel()
    grid=np.c_[xx_,yy_]
    pred=model.predict(grid)
    z=pred.reshape(xx.shape)
    plt.contourf(xx,yy,z)
```

```python
boundary(X,Y,model)
```
